# ㅁ  apply
> apply(function) : Apply function to each object.
***
- 문제1. 'species' 컬럼 데이터의 앞 3글자를 추출하여, 'species_first3'이라는 컬럼에 넣어봅니다.
- 문제2. 'species' 컬럼 데이터의 뒤 3글자를 추출하여, 'species_last3'이라는 컬럼에 넣어봅니다.
<img src='https://drive.google.com/uc?id=1qt1NjLGwpfNeL7ja1dGhiSxGJT-jyOwo'>

```python
df = sns.load_dataset("iris")
df.head()
df['species_first3'] = df['species'].apply(lambda x: x[:3])
df['species_last3'] = df['species'].apply(lambda x: x[3:])
df
```

```python
def first_func(x):
    return x[:3]

def last_func(x):
    return x[3:]

df['species_func_first3'] = df['species'].apply(first_func)
df['species_func_last3'] = df['species'].apply(last_func)
df
```

# ㅁ Combine Data Sets

## 1. merge()
- 문제1. adf, bdf 데이터 프레임을 x1컬럼을 pd.merge()를 사용하여 합치려고 합니다. how 파라메터를 다양하게 주면서 그 의미를 확인해봅니다.
***
<img src='https://drive.google.com/uc?id=1XYG6XRybqhCVBQKbfGC9VmYzzU5FOr-F'>

```python
adf = pd.DataFrame({"x1": ["A", "B", "C"], "x2": [1, 2, 3]})
adf
```

```python
bdf = pd.DataFrame({"x1": ["A", "B", "D"], "x3": ["T", "F", "T"]})
bdf
```

```python
pd.merge(adf, bdf, how='left', on='x1')
```

```python
pd.merge(adf, bdf, how='right', on='x1')
```

```python
pd.merge(adf, bdf, how='inner', on='x1')
```

```python
pd.merge(adf, bdf, how='outer', on='x1')
```

### Q&A) 다음 코드의 의미를 확인해봅니다.
- Method Chaining

```python
ydf = pd.DataFrame({"x1": ["A", "B", "C"], "x2": [1, 2, 3]})
zdf = pd.DataFrame({"x1": ["B", "C", "D"], "x2": [2, 3, 4]})
pd.merge(ydf, zdf, how='outer', indicator=True)\
   .query('_merge == "left_only"') \
   .drop(columns=['_merge'])
```
